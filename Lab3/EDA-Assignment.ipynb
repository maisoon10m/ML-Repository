## 5.Representation and Alnalysis

### Univariate analysis

```python
# Histogram - Total Sales
plt.hist(df['Sales'], bins=30)
plt.title('Distribution of Total Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

Shows how transaction amounts are distributed. You can identify skewness, most common spending range, and presence of extreme high-value purchases.

```python
# Histogram - Unit price
plt.hist(df['Unit price'], bins=30)
plt.title('Distribution of Total Sales')
plt.xlabel('Unit price')
plt.ylabel('Frequency')
plt.show()
```

we can notice : 
- The distribution appears fairly uniform across the price range (10 to 100).
- No strong skewness is observed (not heavily left- or right-skewed).
- Prices are spread evenly, meaning the supermarket offers products across a wide and balanced price range.

## Bivariate Analysis


```python
# Bar Plot (Categorical vs Numerical) - Product line vs Sales 
df.groupby('Product line')['Sales'].mean().plot(kind='bar')
plt.title('Average Sales per Product Line')
plt.ylabel('Average Sales')
plt.show()
```

The chart shows that Home and lifestyle has the highest average sales among all product lines, indicating stronger revenue per transaction in this category. Sports and travel also performs well, with slightly lower but comparable average sales. On the other hand, Fashion accessories records the lowest average sales. Overall, the differences between product lines are relatively small, suggesting that sales are fairly balanced across categories

```python
# Scatter Plot (Numerical vs Numerical) - Unit price vs Sales
plt.scatter(df['Unit price'], df['Sales'])
plt.title('Unit Price vs Sales')
plt.xlabel('Unit Price')
plt.ylabel('Sales')
plt.show()
```

The scatter plot shows a clear positive relationship between Unit Price and Sales, meaning that as the unit price increases, total sales tend to increase as well. The upward linear pattern suggests a strong association between the two variables. The diagonal banding pattern is likely due to different quantity levels, since sales are calculated based on unit price multiplied by quantity.

## Multivariate Analysis (Correlation)

```python
# Heatmap - Unit price & Quantity & Tax 5% & Sales & cogs & gross income & Rating
plt.figure(figsize=(8,6))
sns.heatmap(df[['Unit price','Quantity','Tax 5%','Sales','cogs','gross income','Rating']].corr(),
            annot=True,
            cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

The heatmap shows a very strong positive correlation between Sales, Tax 5%, cogs, and gross income, with correlation values equal to 1. This indicates perfect linear relationships, which is expected because these variables are directly calculated from one another.

Quantity also shows a strong positive correlation (0.71) with Sales-related variables, meaning higher quantities significantly increase total sales and related financial metrics.

Unit price has a moderate positive correlation (0.63) with Sales and related variables, indicating that price also contributes to higher revenue, though slightly less strongly than quantity.

Rating shows almost no correlation with the other variables, suggesting customer ratings are independent of price, quantity, and sales performance.

## Time-Based Analysis

```python
# Convert Date
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# line plot - Daily Sales Trend
daily_sales = df.groupby('Date')['Sales'].sum()

daily_sales.plot()
plt.title('Daily Sales Trend')
plt.ylabel('Total Sales')
plt.show()
```

The daily sales trend shows noticeable fluctuations throughout the observed period, indicating variability in daily revenue. Sales frequently spike above 6,0007,000 on certain days, suggesting peak business periods, while some days drop below 2,000, reflecting lower customer activity. There is no clear consistent upward or downward trend over time; instead, the pattern suggests irregular demand with sharp increases and decreases.

```python
# Line plot - Sales by Hour
hourly_sales = df.groupby('Hour')['Sales'].sum().sort_index()
plt.figure(figsize=(8,5))
plt.plot(hourly_sales.index, hourly_sales.values)
plt.title('Sales by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Total Sales')
plt.xticks(hourly_sales.index)
plt.show()
```

The sales by hour plot shows that revenue varies noticeably throughout the day. Sales are relatively strong during the late morning, dip slightly around midday, and increase again in the early afternoon. A significant peak occurs at 19:00 (7 PM), indicating that early evening is the busiest period for sales. After this peak, sales drop sharply by 20:00, suggesting reduced customer activity toward closing time. Overall, the highest sales activity occurs in the evening hours.
